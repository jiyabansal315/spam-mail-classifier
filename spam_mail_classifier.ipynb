```python
import numpy as np
import pandas as pd
df=pd.read_csv('Mail Data.csv')
```

```python
df
```

```python
# importing all dependencies

from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer # to convert the data into numbers
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# replace the null values with a null string
mail_data = df.where((pd.notnull(df)),'')
mail_data.head()
```

```python
# label spam mail as 0 , ham mail as 1
mail_data['Category']=mail_data['Category'].astype(int)

mail_data.loc[mail_data['Category'] == 'spam', 'Category'] = 0
mail_data.loc[mail_data['Category'] == 'ham' , 'Category'] = 1
```

```python
x=mail_data['Message']
y=mail_data['Category']
```

```python
# split data into training and testing data
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=3)
```

```python
# transform the text data to feature vectors that can be used as input to logistic regression
feature_extraction = TfidfVectorizer(min_df=1,stop_words='english',lowercase=True)
x_train_features=feature_extraction.fit_transform(x_train)
x_test_features=feature_extraction.transform(x_test)
```

```python
model=LogisticRegression()
model.fit(x_train_features,y_train)
print(x_train_features)
```

```python
y_train_pred=model.predict(x_train_features)
y_test_pred=model.predict(x_test_features)
```

```python
accuracy_train=accuracy_score(y_train,y_train_pred)
accuracy_test=accuracy_score(y_test,y_test_pred)
print(accuracy_train,accuracy_test)
```

```python
# predicting using our own input
input_mail=["7 Days Only!Extra Savings For Your Mid-Year Break!Hi jiya bansal,The mid-year break is almost here. It's now time to plan your Singapore city escape.For one week only, enjoy an additional 5% OFF our Advance Purchase rates and unlock even greater value for your upcoming urban holiday getaway."]


input_data_features = feature_extraction.transform(input_mail)
prediction=model.predict(input_data_features)
if (prediction==1):
    print("ham")
else:
    print("spam")
```
